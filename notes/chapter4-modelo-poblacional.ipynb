# Modelo poblacional

```python
import modsim as ms
import matplotlib.pyplot as plt
from pandas import read_html

wiki = 'https://en.wikipedia.org/wiki/Estimates_of_historical_world_population'
```

```python
tablas = read_html(wiki, header=0, index_col=0, decimal='M')
tablas
```

```python
tabla2 = tablas[2]
tabla2.columns = ['census', 'prb', 'un', 'maddison', 'hyde', 'tanton', 'biraben', 'mj', 'thomlinson',
                  'durand', 'clark']
tabla2
```

```python
# Expresamos las variables como miles de millones de personas 
census = tabla2.census/1e9
census
```

```python
un = tabla2.un/1e9
un
```

```python
ms.plot(census, ':', label='US Census')  # : linea de puntos
ms.plot(un, '--', label='UN DESA')  # -- linea de rayas
ms.decorate(title="Poblacion mundial", xlabel="Año", ylabel="Poblacion (miles de millones)")
plt.show()
```

```python
# Primera y ultima fecha :
t_0 = ms.get_first_label(census)
t_end = ms.get_last_label(census)

# Cantidad de anios :
anios = t_end - t_0

# Primera y última población :
p_0 = ms.get_first_value(census)
p_end = ms.get_last_value(census)

# Crecimiento total de la población :
crecimiento_total = p_end - p_0
print(crecimiento_total)
crecimiento_anual = crecimiento_total / anios
print(crecimiento_anual)
```
